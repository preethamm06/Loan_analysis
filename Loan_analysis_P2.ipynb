```python
import pandas as pd
import matplotlib.pyplot as plt
from pyspark.sql.functions import *
from pyspark.sql.functions import desc
# import seaborn as sb


# spark=SparkSession.builder.appName("project2").getOrCreate()
df= spark.read.format("csv").option("header", "true").load("dbfs:/FileStore/shared_uploads/vishwakarmaplc@gmail.com/loan_default_dataset.csv")
dfp=pd.read_csv("/dbfs/FileStore/shared_uploads/vishwakarmaplc@gmail.com/loan_default_dataset.csv")

def cont():
    print("Do You want to continue")
    print("Y.Yes")
    print("N.No ")
    s=input("Enter the value: ").lower()
    if s=="y":
        welcome()
    else:
        print("Thank You")



def desc():
    des=dfp.describe()
    print(des)
    cont()
def countofclient():
    print("Total Number of Clients: ",df.count())
    print("Total Number of Male Clients: ",df.filter(df["Client_Gender"] == "Male").count())
    print("Total Number of Female Clients: ",df.filter(df["Client_Gender"] == "Female").count())
    cont()
    
def ca_gv(): #ca:Credit amount gv:gender vise m:male
    #max credit amount for male client
    df.groupBy("Client_Gender").agg(max("Credit_Amount").alias("Maximum Credit amount")).show()
    #min credit amount for male client
    df.groupBy("Client_Gender").agg(min("Credit_Amount").alias("Minimum Credit amount")).show()
    #Male clients total average credit amount
    df.groupBy("Client_Gender").agg(avg("Credit_Amount").alias("Average Credit amount")).show()
    cont()
    
def ca_ci():  #Client_Income_Type
    #Max credit amount for different income type
    df.groupBy("Client_Income_Type").agg(max("Credit_Amount").alias("Maximum Credit amount")).show()
     #min credit amount for different income type
    df.groupBy("Client_Income_Type").agg(min("Credit_Amount").alias("Minimum Credit amount")).show()
     #avg credit amount for different income type
    df.groupBy("Client_Income_Type").agg(avg("Credit_Amount").alias("Average Credit amount")).show()
    cont()
    
def ca_co(): #ce:Client_Occupation
    #max credit amount for different client ocupations
    df.groupBy("Client_Occupation").agg(max("Credit_Amount").alias("Maximum Credit amount")).show()
     #min credit amount for different client ocupations
    df.groupBy("Client_Occupation").agg(min("Credit_Amount").alias("Minimum Credit amount")).show()
     #avearge credit amount for different client ocupations
    df.groupBy("Client_Occupation").agg(avg("Credit_Amount").alias("Average Credit amount")).show()
    cont()
    
def ca_ce(): #ce: client education   
    #max credit amount for different Client_Education
    df.groupBy("Client_Education").agg(max("Credit_Amount").alias("Maximum Credit amount")).show()
     #min credit amount for different Client_Education
    df.groupBy("Client_Education").agg(min("Credit_Amount").alias("Minimum Credit amount")).show()
     #avearge credit amount for different Client_Education
    df.groupBy("Client_Education").agg(avg("Credit_Amount").alias("Average Credit amount")).show()
    cont()
    
def ca_ms(): #ms:marritial status
    #max credit amount for different Client_Marital_Status
    df.groupBy("Client_Marital_Status").agg(max("Credit_Amount").alias("Maximum Credit amount")).show()
     #min credit amount for different Client_Marital_Status
    df.groupBy("Client_Marital_Status").agg(min("Credit_Amount").alias("Minimum Credit amount")).show()
     #avearge credit amount for different Client_Marital_Status
    df.groupBy("Client_Marital_Status").agg(avg("Credit_Amount").alias("Average Credit amount")).show()
    print("M:Married","D:Divorced","W:Widowed","S:Single")
    cont()
    
#Client income from now onwards
def top_10():
    from pyspark.sql.functions import desc
    #the top 10 clients with the highest Client_Income
    top_10_clients = df.orderBy(desc("Client_Income")).limit(10)
    # top_10_clients.show()
    pan=top_10_clients.toPandas()
    print(pan)
    cont()
    
def ci_gv(): #ci:client income gv:gender vise     
    #Max income for distinct gender
    df.groupBy("Client_Gender").agg(max("Client_Income").alias("Maximum Income amount")).show()
    #min income for distinct gender
    df.groupBy("Client_Gender").agg(min("Client_Income").alias("Minimum Income amount")).show()
    #Max income for distinct gender
    df.groupBy("Client_Gender").agg(max("Client_Income").alias("Average Income amount")).show()
    cont()
    
def ci_ms():
    #max Client_Income for different Client_Marital_Status
    df.groupBy("Client_Marital_Status").agg(max("Client_Income").alias("Maximum Income amount")).show()
     #min Client_Income for different Client_Marital_Status
    df.groupBy("Client_Marital_Status").agg(min("Client_Income").alias("Minimum Income amount")).show()
     #avearge Client_Income for different Client_Marital_Status
    df.groupBy("Client_Marital_Status").agg(avg("Client_Income").alias("Average Income amount")).show()
    print("M:Married","D:Divorced","W:Widowed","S:Single")
    cont()
    
def ci_co():  #co:client occupation
     #max Client_Income for different client ocupations
    df.groupBy("Client_Occupation").agg(max("Client_Income").alias("Maximum Income amount")).show()
     #min Client_Income for different client ocupations
    df.groupBy("Client_Occupation").agg(min("Client_Income").alias("Minimum Income amount")).show()
     #avearge Client_Income for different client ocupations
    df.groupBy("Client_Occupation").agg(avg("Client_Income").alias("Average Income amount")).show()
    cont()
    
def ci_cit(): #Cit:client income type
    #max Client_Income for different Client_Income_Type
    df.groupBy("Client_Income_Type").agg(max("Client_Income").alias("Maximum Income amount")).show()
     #min Client_Income for different Client_Income_Type
    df.groupBy("Client_Income_Type").agg(min("Client_Income").alias("Minimum Income amount")).show()
     #avearge Client_Incomee for Client_Income_Type
    df.groupBy("Client_Income_Type").agg(avg("Client_Income").alias("Average Income amount")).show()
    cont()
    
def ci_ce(): #ce:client education
    #max Client_Income for different Client_Education
    df.groupBy("Client_Education").agg(max("Client_Income").alias("Maximum Income amount")).show()
     #min Client_Income for different Client_Education
    df.groupBy("Client_Education").agg(min("Client_Income").alias("Minimum Income amount")).show()
     #avearge Client_Income for Client_Education
    df.groupBy("Client_Education").agg(avg("Client_Income").alias("Average Income amount")).show()
    cont()
    
    
#Loan annuity
def la_gv(): #la:loan Annuity gv:gender vise
    #Max loan annuity for distinct gender
    df.groupBy("Client_Gender").agg(max("Loan_Annuity").alias("Maximum Loan Annuity")).show()
    #min loan annuity for distinct gender
    df.groupBy("Client_Gender").agg(min("Loan_Annuity").alias("Minimum Loan Annuity")).show()
    #avg loan annuity for distinct gender
    df.groupBy("Client_Gender").agg(avg("Loan_Annuity").alias("Average Loan Annuity")).show()
    cont()
    
def la_co(): #c0:client occupation
    #Max loan annuity for Client_Occupation
    df.groupBy("Client_Occupation").agg(max("Loan_Annuity").alias("Maximum Loan Annuity")).show()
    #min loan annuity for Client_Occupation
    df.groupBy("Client_Occupation").agg(min("Loan_Annuity").alias("Minimum Loan Annuity")).show()
    #avg loan annuity for dClient_Occupation
    df.groupBy("Client_Occupation").agg(avg("Loan_Annuity").alias("Average Loan Annuity")).show()
    cont()
    
def la_ms():
     #Max loan annuity for Client_Marital_Status
    df.groupBy("Client_Marital_Status").agg(max("Loan_Annuity").alias("Maximum Loan Annuity")).show()
    #min loan annuity forClient_Marital_Status
    df.groupBy("Client_Marital_Status").agg(min("Loan_Annuity").alias("Minimum Loan Annuity")).show()
    #avg loan annuity for Client_Marital_Status
    df.groupBy("Client_Marital_Status").agg(avg("Loan_Annuity").alias("Average Loan Annuity")).show()
    print("M:Married","D:Divorced","W:Widowed","S:Single")
    cont()
    
def edu():
    df.groupBy("Client_Education").count().show()
    client_education_counts=df.groupBy("Client_Education").count().toPandas()
    a=int(input("""To see the data in graph
           1.Bar Graph
           2.Pie Chart"""))
    if a==1:
        #Graph
        #client_education_counts=df.groupBy("Client_Education").count().toPandas()
        plt.bar(client_education_counts["Client_Education"],client_education_counts["count"])
        plt.xlabel("Client_Education")
        plt.ylabel("Count")
        plt.title("Distribution of Client Education")
    if a==2:
        #pie
        
        plt.pie(client_education_counts["count"], labels=client_education_counts["Client_Education"], autopct="%1.1f%%")
        plt.title("Proportion of client education counts")
        plt.show()
    cont()
def marritial():
    df.groupBy("Client_Marital_Status").count().show()
    Client_Marital_Status_count=df.groupBy("Client_Marital_Status").count().toPandas()
    print("*","M:Married","D:Divorced","W:Widow","S:Single")
    a=int(input("""To see the data in graph
           1.Bar Graph
           2.Pie Chart"""))
    if a==1:
        #bar graph
        
        plt.bar(Client_Marital_Status_count["Client_Marital_Status"],Client_Marital_Status_count["count"])
        plt.xlabel("Client_Marital_Status")
        plt.ylabel("Count")
        plt.title("Distribution of Client Marital Status")
    if a==2:
        #pie
        plt.pie(Client_Marital_Status_count["count"], labels=Client_Marital_Status_count["Client_Marital_Status"], autopct="%1.1f%%")
        plt.title("Proportion of Client Marital Status count")
        plt.show()
    cont()
    
def gender():
    df.groupBy("Client_Gender").count().show()
    Client_Gender_count=df.groupBy("Client_Gender").count().toPandas()
    a=int(input("""To see the data in graph
           1.Bar Graph
           2.Pie Chart"""))
    if a==1:
        #grpah
        
        plt.bar(Client_Gender_count["Client_Gender"],Client_Gender_count["count"])
        plt.xlabel("Client_Gender")
        plt.ylabel("Count")
        plt.title("Distribution of Client Gender")
    if a==2:
        #pie
        plt.pie(Client_Gender_count["count"], labels=Client_Gender_count["Client_Gender"], autopct="%1.1f%%")
        plt.title("Proportion of Client Gender count")
        plt.show()
    cont()
    
def occupation():
    df.groupBy("Client_Occupation").count().show()
    Client_Occupation_count=df.groupBy("Client_Occupation").count().toPandas()
    a=int(input("""To see the data in graph
           1.Bar Graph
           2.Pie Chart"""))
    if a==1:
        #graph
        
        plt.bar(Client_Occupation_count["Client_Occupation"],Client_Occupation_count["count"])
        plt.xlabel("Client_Occupation")
        plt.ylabel("Count")
        plt.title("Distribution of Client Occupation")
    if a==2:
        
        #pie
        plt.pie(Client_Occupation_count["count"],labels=Client_Occupation_count["Client_Occupation"],autopct="%1.1f%%")
        plt.title("Proportion of client occupation")
        plt.show()    
    cont()
    
def fam_mem():
    df.groupBy("Client_Family_Members").count().show()
    Client_Family_Members_count=df.groupBy("Client_Family_Members").count().toPandas()
    a=int(input("""To see the data in graph
           1.Bar Graph
           2.Pie Chart"""))
    if a==1:
        
        plt.bar(Client_Family_Members_count["Client_Family_Members"],Client_Family_Members_count["count"])
        plt.xlabel("Client Family Members")
        plt.ylabel("Count")
        plt.title("Distribution of Client Family Members")
    if a==2:
        #pie
        plt.pie(Client_Family_Members_count["count"],labels=Client_Family_Members_count["Client_Family_Members"],autopct="%1.1f%%")
        plt.title("Proportion of Client Family Members")
        plt.show()    
    
    cont()

def welcome():
    print("Choose from (1-10)")
    print("1.Desc of dataset")
    print("2.Total Number of client details")
    print("3.Credit amount details")
    print("4.Client Income details")
    print("5.Loan Annuity details")
    print("6.Clients Education info")
    print("7.Clients marritial status")
    print("8.Clients Gender details")
    print("9.Clients occupation")
    print("10.Family members ")
    n=int(input("Enter the value: "))
    if n==1:
        desc()
    elif n==2:
        countofclient()
    elif n==3:
        #client having maximum credit amount
        df.select("Credit_Amount").agg(max("Credit_Amount")).show()
        #client having min credit amount
        df.select("Credit_Amount").agg(min("Credit_Amount")).show()
        #client having avg credit amount
        df.select("Credit_Amount").agg(avg("Credit_Amount")).show()
        print("Choose from (1-5)")
        print("1.Gender wise")
        print("2.Income type vise")
        print("3.Client occupation vise")
        print("4.Client Education vise")
        print("5.Marritial Status vise")
        a=int(input("Select the option: "))
        if a==1:
            ca_gv()
        elif a==2:
            ca_ci()
        elif a==3:
            ca_co()
        elif a==4:
            ca_ce()
        elif a==5:
            ca_ms()
        else:
            print("Enter Proper Number")

    elif n==4:
        #Client having max income
        df.select("Client_Income").agg(max("Client_Income")).show()
        #Client having min income
        df.select("Client_Income").agg(min("Client_Income")).show()
        #Client having avg income
        df.select("Client_Income").agg(avg("Client_Income")).show()
        print("Choose from (1-6)")
        print("1.To view top 10 clients with maximum Income")
        print("2.Gender vise")
        print("3.marritial status")
        print("4.client occupation")
        print("5.client income type")
        print("6.client education ")
        b=int(input("Enter the value: "))
        if b==1:
            top_10()
        elif b==2:
            ci_gv()
        elif b==3:
            ci_ms()
        elif b==4:
            ci_co()
        elif b==5:
            ci_cit()
        elif b==6:
            ci_ce()
        
    elif n==5:
        print("Loan Annuity")
        print("Choose from (1-3)")
        print("1.Gender vise ")
        print("2.Client Occupation ")
        print("3.Cleint marritial status")
        c=int(input("Enter the Value: "))
        if c==1:
            la_gv()
        elif c==2:
            la_co()
        elif c==3:
            la_ms()
    elif n==6:
        edu()
    elif n==7:
        marritial()
    elif n==8:
        gender()
    elif n==9:
        occupation()
    elif n==10:
        fam_mem()

welcome()
```
